```python
from PIL import Image, ImageDraw
from face_detect import detect_faces, crop
import warnings
warnings.filterwarnings("ignore")
```

```python
img = Image.open('1.jpg').convert('RGB')
draw = ImageDraw.Draw(img)
```

```python
%%time
faces, ldmrks = detect_faces(img)
```

```python
crop(img, faces)
```

```python
imgs = []
for line in faces:
    line_ = line.tolist()
    tmp = img.crop(line_[:4])
    imgs.append(tmp)
```

```python
imgs[1].save('test.jpg')
```

```python
for line in faces:
    line_ = line.tolist()
    draw.rectangle(line_[:4], outline='red')
    draw.text(line_[:2], text='{:.2f}'.format(line_[-1]*100) ,fill='red')
```
